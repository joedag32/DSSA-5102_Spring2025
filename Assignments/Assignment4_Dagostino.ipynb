Uh oh...turns out it wasn't just one person behind the crime...who is this additional person involved?<br>
We have some clues, her height, hair color and car model. Also that she attended a concert 3 times.

```python
#Remember to write our your query in the comments!
#You can complete this in a single query or multiple queries.
with engine.connect() as connection: # setup our connection
    query = text("""SELECT *
                    FROM drivers_license
                    INNER JOIN person ON drivers_license.id = person.license_id
                    WHERE height BETWEEN 65 AND 67 
                    AND hair_color = 'red'
                    AND gender = 'female'
                    AND car_make = 'Tesla'
                    AND car_model = 'Model S'
                 """)                           # ok, first I was going to do multiple queries, but I figured let's try this out with a join
                                                # select all records from both drivers_license and person tables and join them together where 
                                                # drivers_license id is equal to person license_id are foreign keys that connect the 2 tables
                                                # then filter out records where height is between 65 and 67 inches,hair_color is "red",
                                                # gender equals "female", car_make equals "tesla", and car_model equals "Model S"
    accomplice = pd.read_sql(query, connection) # use pandas to read the sql query with the connection to the db

accomplice
# got 3 results and the inner join worked!
```

```python
#Let's find the killer!
#Remember to write our your query in the comments!
with engine.connect() as connection: # setup our connection
    query = text("""SELECT *
                    FROM facebook_event_checkin
                    INNER JOIN person ON facebook_event_checkin.person_id = person.id
                    WHERE event_name = 'SQL Symphony Concert'
                    AND person_id IN (78881, 90700, 99716)
                 """)                                 # let's do another join, select all records from facebook_event_checkin and person tables
                                                      # and join them together where their person.id columns match which happen to be foreigh
                                                      # keys, then gather records between the 2 where event_name equals "SQL Symphony Concert"
                                                      # and the person_id is either 78881, 90700 or 99716
    concert_attendee = pd.read_sql(query, connection) # use pandas to read the sql query with the connection to the db

concert_attendee
```
