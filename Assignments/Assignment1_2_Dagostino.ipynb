<h4>Data Cleaning</h4>

The Olympic data was filtered to just USA athletes in Assignment 1, so right off the bat I narrowed the data down to USA athletes.

Make all column names lowercase, with no spaces or special characters:

```python
# Make column names lowercase to play well with SQL
usa_athletes_df.columns = usa_athletes_df.columns.str.lower()
print(usa_athletes_df.columns) # output column names to test if lowercase
```

```python
# Happy Cleaning :) 
# todo list: remove duplicates, eliminate whitespace, null vaules, remove any columns?
```

```python
# see if there's any duplicate rows
duplicateRows = usa_athletes_df[usa_athletes_df.duplicated()]
display(duplicateRows) #display some of the duplicate rows to be sure
```

After inspection there are 249 duplicate rows, so lets clean that up

```python
# Remove dumplicates with drop_duplicates()
usa_athletes_clean_df = usa_athletes_df.drop_duplicates()
display(usa_athletes_clean_df) # display to give it another look
```

Remove leading and trailing whitespace

```python
# run strip() on the name and event columns to remove any leading and trailing whitespace
usa_athletes_clean_df['name'].str.strip()
usa_athletes_clean_df['event'].str.strip()
display(usa_athletes_clean_df) # display to give it another look
```

Let's remove unnecessary columns

```python
# first double check the unique values in noc
print(usa_athletes_clean_df['noc'].unique())
```

```python
# going to delete this column since they're all USA
usa_athletes_clean_df = usa_athletes_clean_df.drop(['noc'], axis=1) # call drop() on the noc column to remove it
print(usa_athletes_clean_df.columns) # output column names to make sure noc was dropped
```

```python
# delete the games column, it's redundant with the year and season columns
usa_athletes_clean_df = usa_athletes_clean_df.drop(['games'], axis=1) # call drop() on the games column to remove it
print(usa_athletes_clean_df.columns) # output column names to make sure noc was dropped
```
